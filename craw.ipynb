```python

from bs4 import BeautifulSoup
import urllib.request
import concurrent.futures as cf
import cssutils
```

```python
def openBrowser(url):
   page = urllib.request.urlopen(url)
   soup = BeautifulSoup(page, 'html.parser')
   return soup
```

```python
soup = openBrowser('https://mia.vn/cam-nang-du-lich')
data = soup.find_all('a',class_ = 'destination-item')
links = [ ("https://mia.vn/" + i.get('href')) for i in data]
```

```python
def remove_item_line(e):
   return list(filter(lambda x : x != '\n',e))
```

```python
import re
def removeAllStr(e):
   return re.sub('\s+', ' ', e)
```

```python

def crawData(e):
   item = remove_item_line(e)[1:]
   item = list(e.children)
   item = remove_item_line(e)[1:]
   item_img = item[0].findChildren()[0].contents[1].get('data-original')
   item_content = remove_item_line(list(remove_item_line(list(item[1].children))[0].children))
   return {
      "imgUrl" : "https:" + item_img,
      "tiltle-content" : item_content[0].text,
      "content" : removeAllStr(item_content[1].text)
   }

```

```python
import numpy as np
import functools as fu
events = [f'https://lug.vn/su-kien?page={i}' for i in range(1,4)]
def drawAllData(envent):
   html_event =  openBrowser(envent)
   list_box  =  list(html_event.find(class_= 'list-blog').children)
   list_box =  remove_item_line(list_box)
   with cf.ThreadPoolExecutor() as exe:
      return list(exe.map(crawData,list_box))
with cf.ThreadPoolExecutor() as exe:
      data = list(exe.map(drawAllData,events))
data
import json
with open("./assets/data/sukien.json", "w") as outfile:
   json.dump(data, outfile)
# len(fu.reduce(lambda x,y : x + y,data))













```

```python

def getLink(links):
   data = openBrowser(links)
   info = data.find(class_='dest-info-box').findChildren()
   return {
     "name" : info[2].findChildren()[0].get_text(),
     "description" : info[2].findChildren()[1].get_text(),
     "imgUrl" : cssutils.parseStyle(info[1].get('style'))['background-image'].replace('url(', '').replace(')', '')
   }
```

```python

with cf.ThreadPoolExecutor() as exe:
    final_data = list(exe.map(getLink,links))
final_data
```

```python
import json
with open("./assets/data/chitietcamnang.json", "w") as outfile:
   json.dump(final_data, outfile)
```
